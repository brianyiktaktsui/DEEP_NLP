```python
selectedAttrib='description'
timeStamp='1535393121.334881'
```

```python
import jsonlines
myL=[]
with jsonlines.open( "./Data/validation_data_curation/Deep NLP (3).json") as reader:
    for obj in reader:
        orig_text=obj['content']
        #if orig_text=='Time course experiment wild mice moved lab':
        for labeledDict in obj['annotation']:
            for myLabel in labeledDict['label']:
                #print (len(labeledDict['points']))
                tmpS=pd.Series( labeledDict['points'][0])
                tmpS.loc['content']=orig_text
                tmpS.loc['label']=myLabel
                myL.append(tmpS)
            
```

```python
mergedDf_curated=pd.concat(myL,axis=1).T
```

### curate overlap against curation
    

```python
mergedDf_curated.head()
```

```python
mergedDf_curated['text_len']=mergedDf_curated['text'].str.len()
```

```python
curatedDf_interval=mergedDf_curated[mergedDf_curated.label!='UNKNOWN'].drop_duplicates(
    ['content','label']).set_index(['content','label'])[['start','end','text_len']]#.index
```

```python
#curatedDf_interval
```

```python
hitDf['text_len']=hitDf['token_len']
```

```python
hitDf['ith_char_end_pos']=hitDf['ith_char_pos']+hitDf['token_len']-1
```

```python
hitDf_interval=hitDf.drop_duplicates(['orig_text','predicted']
                                    ).set_index(['orig_text','predicted'])[['ith_char_pos','ith_char_end_pos','text_len']]
```

```python
#hitDf_interval['']
```

```python
hitDf_interval.columns=['start','end','text_len']
```

```python
curatedDf_interval.head()
```

```python
hitDf_interval.head()
```

```python
mergedDf=pd.concat([curatedDf_interval,hitDf_interval],keys=['curated','predicted'],axis=1).dropna().astype(float)
```

```python
"""
def overlap(min1, max1, min2, max2):
    return max(0, min(max1, max2) - max(min1, min2))

"""

def overlap(tmpS):
    min1=tmpS.loc['predicted']['start']
    max1=tmpS.loc['predicted']['end']
    
    min2=tmpS.loc['curated']['start']
    max2=tmpS.loc['curated']['end']
    return max(0, min(max1, max2) - max(min1, min2))
    #return max(0, min(max1, max2) - max(min1, min2))
```

```python
mergedDf['word count']=mergedDf.index.get_level_values(0).str.len()
```

```python
#mergedDf
```

```python
mergedDf['% of curated covered in predicted']=(mergedDf.apply(overlap,axis=1)/mergedDf['curated']['text_len'])
ax=mergedDf['% of curated covered in predicted'].hist(bins=8)
ax.set_ylabel('% of curated intervals')
ax.set_xlabel('% of curated covered in predicted')
```
