```python
import pandas as pd

selectedAttrib='description'
timeStamp='1535393121.334881'

hitDf=pd.read_pickle('./Data/validation_data/validataion_prediction_{}.{}.pickle'.format(selectedAttrib,timeStamp))
```

```python
hitDf.head()
```

```python
import jsonlines
myL=[]
with jsonlines.open( "./Data/validation_data_curation/Deep NLP description.json") as reader:
    for obj in reader:
        orig_text=obj['content']
        #if orig_text=='Time course experiment wild mice moved lab':
        for labeledDict in obj['annotation']:
            for myLabel in labeledDict['label']:
                #print (len(labeledDict['points']))
                tmpS=pd.Series( labeledDict['points'][0])
                tmpS.loc['content']=orig_text
                tmpS.loc['label']=myLabel
                myL.append(tmpS)
            
```

```python
mergedDf_curated=pd.concat(myL,axis=1).T
```

### curate overlap against curation
    

```python
subCuratedDf=mergedDf_curated[mergedDf_curated['label']!='UNKNOWN']

membership_curated_df=pd.Series(index=subCuratedDf.drop_duplicates(['content','label']).set_index(['content','label']).index).fillna(1).unstack().fillna(0)
```

```python
membership_curated_df.to_pickle('./Data/validation_data_curation/Deep NLP (3).pickle')
```

```python
membership_predicted_df=pd.Series(index=hitDf.drop_duplicates(['orig_text','predicted']).set_index(['orig_text','predicted']).index).fillna(1).unstack().fillna(0)
```

```python
#(membership_curated_df==membership_curated_df).mean()
```

```python
mergedDf=pd.concat([membership_predicted_df,membership_curated_df],axis=1,keys=['predicted','curated']).dropna()
```

```python
#mergedDf['curated']
```

```python
#(mergedDf['predicted']==mergedDf['curated']).mean().mean()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics

predActualDf=mergedDf.stack()
```

```python
predActualDf.head()
```

```python
tn, fp, fn, tp=metrics.confusion_matrix (predActualDf['curated'],predActualDf['predicted']).ravel()#mergedDf['metamap']
myDict={}
myDict['specificity'] = tn / (tn+fp)
myDict['sensitivity']=tp/(tp+fn)
myDict['precision']=tp/(tp+fp)
myDict['recall']=tn / (tn+fp)

myDict['F1']=2/(1/myDict['recall']+1/myDict['precision'])
pd.Series(myDict).to_frame(name='our NER')
```
