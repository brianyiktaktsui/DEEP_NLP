```python

semanticToEntityS=pd.Series({'Anatomy': 'Cell type/tissue',
 'Disorders': 'Disease state',
 'Genes & Molecular Sequences': 'Genotype',
 'Geographic Areas': 'Geographical Location',
 'Living Beings': 'Species',
 'Procedures': 'Treatment and conditions'})
```

```python
membership_predicted_df_metamap=metamapMembershipDf.loc[:,semanticToEntityS.index].copy()
membership_predicted_df_metamap.columns=semanticToEntityS.values
#membership_predicted_df_metamap.columns=semanticToEntityS[metamapMembershipDf_sub.columns].values
#membership_predicted_df_metamap
#semanticToEntityS
```

```python
mergedDf=pd.concat([membership_predicted_df_metamap,membership_predicted_df_mine],axis=1,keys=['metamap','curated']).dropna()
```

```python
from sklearn.metrics import confusion_matrix



from sklearn import metrics

predActualDf=mergedDf.stack()
```

```python
tn, fp, fn, tp=metrics.confusion_matrix (predActualDf['curated'],predActualDf['metamap']).ravel()#mergedDf['metamap']
myDict={}
myDict['specificity'] = tn / (tn+fp)
myDict['sensitivity']=tp/(tp+fn)
myDict['precision']=tp/(tp+fp)
myDict['recall']=tn / (tn+fp)
myDict['F1']=2/(1/myDict['recall']+1/myDict['precision'])
pd.Series(myDict).to_frame(name='metamap')
```

```python
print ('accuracy:',(mergedDf['metamap']==mergedDf['predicted']).mean().mean())
```
